```python
start_date = '2010-01-01'
end_date = '2023-09-01'
mask_date = (df['update_date'] >= start_date) & (df['update_date'] <= end_date)
```

```python
mask_categories=df['categories'].str.contains('cond-mat')
mask_categories_exclude=(~df['categories'].str.contains('cond-mat.soft') & ~df['categories'].str.contains('cond-mat.stat-mech'))
```

```python
df_cmp=df[mask_date & mask_categories & mask_categories_exclude]
df_cmp['abstract']=df_cmp['abstract'].str.lower()
```

```python
mask_HF = (df_cmp['abstract'].str.contains('hartree') | df_cmp['abstract'].str.contains('fock') | df_cmp['abstract'].str.contains('mean-field'))
```

```python
mask_exclude_nonfermion= (
  ~df_cmp['abstract'].str.contains('parton') 
& ~df_cmp['abstract'].str.contains('boson') 
& ~df_cmp['abstract'].str.contains('bose') 
)
mask_exclude_nonnormal=(
  ~df_cmp['abstract'].str.contains('spin liquid') 
& ~df_cmp['abstract'].str.contains('spin-glass') 
& ~df_cmp['abstract'].str.contains('superfluid') 
& ~df_cmp['abstract'].str.contains('superconductivity') 
& ~df_cmp['abstract'].str.contains('bose-einstein') 
& ~df_cmp['abstract'].str.contains('phonon') 
& ~df_cmp['abstract'].str.contains('photon') 
& ~df_cmp['abstract'].str.contains('bose-hubbard') 
& ~df_cmp['abstract'].str.contains('superconducting') 
)
mask_exclude_nonHF=(
  ~df_cmp['abstract'].str.contains('monte') 
& ~df_cmp['abstract'].str.contains('carlo') 
& ~df_cmp['abstract'].str.contains('dft') 
& ~df_cmp['abstract'].str.contains('first-principles') 
& ~df_cmp['abstract'].str.contains('density functional theory') 
& ~df_cmp['abstract'].str.contains('renormalization group') 
& ~df_cmp['abstract'].str.contains('renormalization') 
& ~df_cmp['abstract'].str.contains('coarse-grained') 
& ~df_cmp['abstract'].str.contains('ginzburg-landau') 
& ~df_cmp['abstract'].str.contains('exact diagonalization') 
& ~df_cmp['abstract'].str.contains('dmft') 
& ~df_cmp['abstract'].str.contains('dynamical mean-field') 
& ~df_cmp['abstract'].str.contains('random matrix') 
)
mask_exclude_topic=(
  ~df_cmp['abstract'].str.contains('spin-orbital')
&  ~df_cmp['abstract'].str.contains('landau level')
&  ~df_cmp['abstract'].str.contains('real-space hartree-fock')
)

mask_exclude= mask_exclude_nonfermion & mask_exclude_nonnormal & mask_exclude_nonHF & mask_exclude_topic
# gradually reduce from fermion /topic / method/ 
```
